```python
def recurse(tree_,node, depth,columns):
    dependencies = []
    val = tree_.feature[node]
    len_x = numColonnes - len(labels)
    if  val != tree._tree.TREE_UNDEFINED:
        if val >= len_x :
            name = columns[val]
            dependencies.append({'class':name,'depth':depth})
        dependencies = dependencies + recurse(tree_,tree_.children_left[node], depth + 1,columns)
        dependencies = dependencies + recurse(tree_,tree_.children_right[node], depth + 1,columns)
    return dependencies
```
