```python
from multilabel3rc import ML3RC
import arff, numpy as np
from arff2pandas import a2p
attribute_number = 72
class_number=6

with open('emotions/emotions-train.arff','rt') as f:
    df_train = a2p.load(f)
with open('emotions/emotions-test.arff','rt') as f:
    df_test = a2p.load(f)
    
labels = df_train.columns[attribute_number:]
XY_train = df_train
y_train = df_train[labels]
XY_test = df_test
y_test = df_test[labels]
# df=df.rename(columns={df.columns[0]:"protein"})
# df=df.replace('NO',0)
# df=df.replace('YES',1)
```

```python
model = ML3RC()
model.fitdata(XY_train,y_train)
```

```python
p=model.Predict(XY_test.iloc[:,:attribute_number])
```

```python
score=model.accuracyScore(y_test,p)
```
